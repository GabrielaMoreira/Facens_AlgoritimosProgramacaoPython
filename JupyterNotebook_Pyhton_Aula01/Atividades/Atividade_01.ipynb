## Escreva uma função que receba três parâmetros (a, b e c) inteiros e retorne o maior valor entre os argumentos passados

```python
def max_value(a, b, c):
    if a < b:
        if b < c:
            return c
        else:
            return b
    else:
        if a < c:
            return c
        else:
            return a
```

```python
assert max_value(1, 2, 3) == 3
assert max_value(5, 2, 3) == 5
assert max_value(5, 12, 3) == 12
```
